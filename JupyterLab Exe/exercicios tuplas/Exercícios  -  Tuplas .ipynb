# Exercício 10: Jogo de Cartas
### Crie um jogo de cartas em que dois jogadores jogam alternadamente. Cada jogador recebe uma mão de cartas (uma tupla de strings). Implemente a lógica de jogo, como verificar vitória, derrota ou empate, e exiba o resultado após cada rodada.

### Esses exercícios de nível médio a difícil envolvem conceitos mais avançados de manipulação de tuplas e resolução de problemas. Eles podem ajudar a aprofundar sua compreensão das tuplas e como elas podem ser aplicadas em situações mais complexas.

```python
import random
# pondo na sequencia de maior ao menor
valores = ['4', '5', '6', '7', 'Q', 'J', 'K', 'Ás', '2', '3']
nipes = ['Paus', 'Copas', 'Espadas', 'Ouros']

# lista e tupla para a criação do baralho
baralhos = []
baralho = ()

# mão de cada jogador (3 cartas que cada um possui)
jogador1 = []
jogador2 = []

# varaiveis para pontuação de cada jogador
pontos1 = 0
pontos2 = 0

# variaveis de cartas
carta1 = ''
carta2 = ''

# fazendo um looping para cada valor dentro do nipes
for i in range (0, len(nipes), 1) :
    nipe = nipes[i]
    # looping para pegar o valor da carta dentro de valores, adicionando essas duas variaveis dentro da tupla baralho para ai sim por na lista baralhos
    for x in range (0, len(valores), 1) :
        valor = valores[x]
        baralho = nipe, valor
        baralhos.append(baralho)
        #print (baralho)

# destribuindo cartas
for i in range (0, 3, 1) :
    # tirando para o primeiro jogador e retirando a carta dele do baralhos
    tirada1 = random.choice(baralhos)
    jogador1.append(tirada1)
    baralhos.remove(tirada1)
    # tirando para o primeiro jogador e retirando a carta dele do baralhos
    tirada2 = random.choice(baralhos)
    jogador2.append(tirada2)
    baralhos.remove(tirada2)


# criando a manilha da partida
vira = random.choice(baralhos)
baralhos.remove(vira)

# caso a manilha tenha um valor fora da lista, retorna sendo o primeiro (0) indice
manilha = valores.index(vira[1]) + 1
if manilha > len(valores) :
    manilha = 0

# mostrando a manilha, removendo ela da lista e adicionando ela no final. Assim quando ivertemos a lista o indice da manilha será o primeiro (0), fazendo com que seja a maior do jogo
print (f'Manilha: {valores[manilha]} (Paus > Copas > Espadas > Ouros)')
addManilha = valores.pop(manilha)
valores.append(addManilha)
valores.reverse()

# fazendo com que o programa rode enquato nenhum dos jogadores tiver 2 pontos
# fazendo as 3 partidas   
while pontos1 <= 1 and pontos2 <= 1 :
        # mostrando as cartas do primeiro jogador
        if pontos1 < 2 or pontos2 < 2 :
            print (f'-- Cartas do Jogador 1 --')
            for i in range (0, len(jogador1), 1):
                print (f'{i + 1}º {jogador1[i]}')

            # mostrando as cartas do segundo jogador jogador
            print (f'-- Cartas do Jogador 2 --')
            for i in range (0, len(jogador2), 1):
                print (f'{i + 1}º {jogador2[i]}')


        # ----- JOGADOR 1 -----
        # pedindo para o jogador 1 escolher uma carta dele para jogar
        jogada1 = round(int (input ('(Jogador 1)Digite o numero da carta que deseja jogar.')))
        jogada1 = jogada1 - 1
        carta1 = jogador1[jogada1]
        print (f'Jogador 1 jogou {carta1}')
        jogador1.remove(carta1)
        valor1 = valores.index(carta1[1])
        nipe1 = nipes.index(carta1[0])

        # ----- JOGADOR 2 -----
        # pedindo para o jogador 2 escolher uma carta dele para jogar
        jogada2 = round(int (input ('(Jogador 2)Digite o numero da carta que deseja jogar.')))
        jogada2 = jogada2 - 1
        carta2 = jogador2[jogada2]
        print (f'Jogador 2 jogou {carta2}')
        jogador2.remove(carta2)
        valor2 = valores.index(carta2[1])
        nipe2 = nipes.index(carta2[0])


        #  comparando as cartas e vendo qual a maior
        if valor1 < valor2  :
            print ('jogador 1 ganha 1 ponto.')
            pontos1 += 1

        elif nipe1 < nipe2 :
            print ('jogador 1 ganha 1 ponto.')
            pontos1 += 1

        else :
            print ('Jogador 2 ganhou 1 ponto.')
            pontos2 += 1

# mostrando qual dos jogadores foi o vencedor    
if pontos1 == 2 :
    print ('Jogador 1 é o grande vencedor.')
        
else :
    print ('Jogador 2 é o grande vencedor.')
            
         
        # completo

    
```
